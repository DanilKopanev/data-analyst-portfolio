<font color='#777778'>

В исследовании используются два датасета о заведениях общественного питания Москвы.

**Файл `/datasets/rest_info.csv`** содержит информацию о самих заведениях:  

- `name` — название заведения.  
- `address` — адрес заведения.  
- `district` — административный район Москвы (например, Центральный административный округ).  
- `category` — категория заведения (например, «кафе», «пиццерия», «кофейня» и т.п.).  
- `hours` — информация о днях и часах работы.  
- `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (от 0 до 5.0).  
- `chain` — признак сетевого заведения: `0` — не сетевое, `1` — сетевое (для небольших сетей возможны ошибки).  
- `seats` — количество посадочных мест.  

**Файл `/datasets/rest_price.csv`** содержит информацию о ценовом уровне заведений:  

- `price` — категория цен (например, «средние», «ниже среднего», «выше среднего» и т.д.).  
- `avg_bill` — текстовое описание среднего чека или цены (диапазон или одно значение), например:  
  - «Средний счёт: 1000–1500 ₽»  
  - «Цена чашки капучино: 130–220 ₽»  
  - «Цена бокала пива: 400–600 ₽» и др.  
- `middle_avg_bill` — численная оценка среднего чека для строк `avg_bill`, начинающихся с «Средний счёт»:  
  - при диапазоне из двух значений — медиана этого диапазона,  
  - при одном значении — само это значение,  
  - в остальных случаях — пропуск.  
- `middle_coffee_cup` — численная оценка цены одной чашки капучино для строк `avg_bill`, начинающихся с «Цена одной чашки капучино»:  
  - при диапазоне из двух значений — медиана этого диапазона,  
  - при одном значении — само это значение,  
  - в остальных случаях — пропуск.

```python
def get_dup_groups(df, col_norm):
    counts = df.groupby(col_norm)[col_norm].count()
    dup_vals = counts[counts > 1].index
    return df[df[col_norm].isin(dup_vals)].sort_values(col_norm) # проверяем входит ли значение в набор

name_groups = get_dup_groups(df, 'name_norm')
addr_groups = get_dup_groups(df, 'address_norm')
```

```python
# Создаём функцию для подсчёта количества объектов и долю каждого значения
def counts_with_share(df, column, rename_map=None):
    s = df[column]
    if rename_map is not None:
        s = s.map(rename_map) # используем ".map()" для заменены значения по словарю
    

    counts = (
        s.value_counts()
        .reset_index()
    )
    counts.columns = [column, 'count']
    counts['share'] = (counts['count'] / counts['count'].sum() * 100).round()
    return counts


chain_counts = counts_with_share(df,'chain',rename_map={0: 'несетевое', 1: 'сетевое'})

chain_counts
```

```python
# Создали функцию для горизонтального barplot
def plot_barh_counts(data, cat_col, count_col='count', title='', xlabel='', ylabel=''):
    order = data.sort_values(count_col, ascending=False)[cat_col]
    
    plt.figure()
    sns.barplot(
        data=data,
        y=cat_col,
        x=count_col,
        palette='Blues_r',
        order=order
    )
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()
```

```python
plot_barh_counts( # возможно здесь подошла бы круговоая диаграмма
    chain_counts,
    cat_col='chain',
    title='Соотношение сетевых и несетевых заведений',
    xlabel='Число заведений',
    ylabel='Тип заведения'
)
```

```python
# Посчитали количество по категории и признаку сети
chain_by_cat = (df.groupby(['category', 'chain'])
    .size() # проверял с .count() - более громоздко, так как нужно обращаться к столбцу name
    .reset_index(name='count')
)

# Посчитали суммарное количество заведений в категории
total_by_cat = (chain_by_cat
    .groupby('category')['count']
    .sum()
    .rename('total_count')
)

chain_by_cat = chain_by_cat.merge(total_by_cat, on='category') # соединили таблицы

# Оставили только сетевые
chain_only = chain_by_cat[chain_by_cat['chain'] == 1].copy() # чтобы не было ошибки

# Посчитали долю сетевых внутри категории
chain_only['share_chain'] = (chain_only['count'] / chain_only['total_count'] * 100).round()

top_chain_categories = (chain_only.sort_values('share_chain', ascending=False))

top_chain_categories[['category', 'share_chain']]
```

Изучили, с какими данными показывают самую сильную корреляцию рейтинги заведений. Выберали самую сильную связь и проверили её.</font>

```python
# В матрицу корреляций включили числовые и категориальные признаки
# без экстремальной доли пропусков
corr_df = df[[
    'rating',
    'category',
    'district',
    'chain',
    'seats',
    'price',
    'is_24_7'
]].copy()

# По условию нужно взять категориальные признаки, изучил .factorize
cat_cols = ['category', 'district', 'price']

for col in cat_cols:
    corr_df[col], _ = pd.factorize(corr_df[col])
```

```python
# Матрица корреляций рейтинга с указанными признаками
corr_matrix = corr_df.corr()

plt.figure()
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    vmin=-1,
    vmax=1
)
plt.title('Матрица корреляций рейтинга и признаков заведения')
plt.tight_layout()
plt.show()
```

В принятой «правиле большого пальца» шкале \(|r| < 0.3\) относят к слабой корреляции, поэтому ни один из этих факторов сам по себе не объясняет рейтинг заведений, а заметной линейной зависимости не наблюдается.
- Дополнительно видно, что между ценовой категорией и районом (`price` и `district`) тоже есть слабая отрицательная связь (около −0.13), то есть более дорогие заведения чуть чаще сконцентрированы в определённых округах, но и эта зависимость остаётся умеренной.

### Рекомендации для заказчика

<div style="background-color: #AEBCC4; padding: 10px; border-radius: 4px;">
